```python
import pandas as pd
Location = 'titanic3.xls'
df = pd.read_excel(Location)
df.head()
```

```python
df.describe()
```

```python
categorical = df.dtypes[df.dtypes == "object"].index
print(categorical)
df[categorical].describe()
```

```python
#since cabin,boat,home.dest,body have many missing numbers, i shall remove these columns.
del df["body"]
df.head()
```

```python
del df["cabin"]
df.head()
```

```python
del df["boat"]
df.head()
```

```python
del df["home.dest"]
df.head()
```

```python
def score_to_numeric(x):
    if x=='female':
        return 1
    if x=='male':
        return 0
df['gender']=df['sex'].apply(score_to_numeric)
df.head()
```

```python
import numpy as np
df['isYoung'] =np.where(df['age']<18, '1', '0')
df.head(10)
```

```python
df["Family_Size"]= df["sibsp"]+df["parch"]
df.head()
```

```python
pd.value_counts(df['gender'])
#35.6% on board were female, while 64.4% were male.
```

```python
pd.value_counts(df['isYoung'])
#Only 12% of the passengers were below 18 years
```

```python
pd.value_counts(df['Family_Size'])
# 60% had no family, while 40% had.
```

```python
pd.value_counts(df['survived'])
# 38% of all passengers survived.
```

```python
pd.crosstab(df.isYoung, df.survived, margins = True)
#53% of young passengers survived, while only 36% of adults survived.
# The young passengers had a higher chance of survival.
```

```python
pd.crosstab(df.gender, df.survived, margins = True)
# 73% of females survived, while 19% of males survived
# Females were therefore more likely to survive
```

```python
pd.crosstab(df.pclass, df.survived, margins = True)
# 62% of those in First class survived, 43% of those in second class survived and 26% of those in third class survived
# Therefore survival was highest for those in First class and lowest in third class.
```

```python
pd.value_counts(df['embarked'])
```

```python
df["age"].fillna(df["age"].mean(), inplace=True)
df.head()
```

```python
df.corr()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula = 'survived~pclass + age + fare + gender + Family_Size', data=df).fit()
result.summary()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula = 'survived~pclass + age + fare + gender + Family_Size -1', data=df).fit()
result.summary()
```

```python
# Adj.R-squared of 0.557 is relatively low even after removing the intercept, therefore only about 55 % of the data variation can be explained by the model
# All P values are significant
# The larger the family size, the less the chance of survival
# gender has a strong positive relationship, whereby females were more likely to survive.
```
